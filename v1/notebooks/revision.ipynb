## Stock price prediction 

```python
### importing the important libraries
import pandas as pd
import numpy as np
import seaborn as sns
import pandas_datareader as pdr
import matplotlib.pyplot as plt
%matplotlib inline
import os
from dotenv import load_dotenv
```

```python
### dealing with the api keys
load_dotenv('keys.txt')
key = os.getenv('key')
```

```python
### loading the stock price by using the api and storing in form of a data frame
stock_price_apple = pdr.get_data_tiingo('AAPL', api_key = key)
stock_price_apple
```

```python
### saving as the CSV
stock_price_apple = stock_price_apple.to_csv('../stocks/apple-18-2.csv')
```

```python
### reading the file again
df_apple = pd.read_csv('../stocks/apple-18-2.csv')
df_apple.head()
```

```python
### checking up on the shape
df_apple.shape
```

```python
### we will be targeting only the 'close' price aka 'close' feature only
df_apple_close_price = df_apple[['close']].values
df_apple_close_price
```

```python
### checking the type of the df_apple_close_price
type(df_apple_close_price)
```

```python
### checking the shape of the df_apple_close_price
len(df_apple_close_price)
```

```python
### plotting the trend
plt.plot(df_apple_close_price)
```

```python
### applying the minmax scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))
df_apple_close_price = scaler.fit_transform(df_apple_close_price)
```

```python
### checking out the things
df_apple_close_price
```

```python
### checking the shape of the scaled data
len(df_apple_close_price)
```

```python
### making the sequences
X, y = [], []
days = 90

for i in range (len(df_apple_close_price) - days) : 

    X.append(df_apple_close_price[i : i + days])
    y.append(df_apple_close_price[i + days])

X = np.array(X)
y = np.array(y)
```

```python
### checking the shape of the sequences
print(f"{X.shape} :: {y.shape}")
```

```python
### train and test data (80-20) split
test_size = int(len(df_apple_close_price) * 0.80)
X_train, X_test = X[:test_size], X[test_size:]
y_train, y_test = y[:test_size], y[test_size:]
```

```python
### checking out the shape of the train and test data
print(f"X_train : {X_train.shape} :: X_test : {X_test.shape} \ny_train : {y_train.shape} :: y_test : {y_test.shape}")
```

```python
### importing the libraries from tensorflow for the model building
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, Dense, Dropout, LSTM
```

```python
### building out the model (sequential type)
model = Sequential([
    LSTM(50, return_sequences = True, input_shape = (days, 1)),
    Dropout(0.33),
    Bidirectional(LSTM(100)),
    Dropout(0.33),
    Dense(1, activation = 'linear')
])
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['mse'])
```

```python
### checking out the model summary
model.summary()
```

```python
### training the model with some batch size and epochs
model.fit(X_train, y_train, epochs = 20, batch_size = 32, verbose = 1)
```

```python
### predictions on the X_test
y_pred = model.predict(X_test)
```

```python
### checking y_pred
y_pred
```

```python
### rescaling the y_pred and y_test
y_pred = scaler.inverse_transform(y_pred)
y_test = scaler.inverse_transform(y_test)
```

```python
### checking the inversly transformed
y_pred
```

```python
### checking out the y_test
y_test
```

```python
### for the comparison plotting the graphs and the metrices
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error
print(f"MAE : {mean_absolute_error(y_test, y_pred)}")
print(f"MSE : {mean_squared_error(y_test, y_pred)}")
print(f"RMSE : {root_mean_squared_error(y_test, y_pred)}")
```

```python
### plotting out the things
plt.figure(figsize = (8, 4))
plt.subplot(1, 2, 1)
plt.plot(y_pred)
plt.plot(y_test)
plt.subplot(1, 2, 2)
plt.scatter(y_pred, y_test)
plt.scatter(y_test, y_pred, edgecolors = 'r')
plt.show()
```
